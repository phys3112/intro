# Comments on PHYS3112 experimental uncertainties questions

This contains discussion of the google form used to explore experimental uncertainties for PHYS3112.

## Measurements with the 1000m ruler

A reasonable estimate of the length of the rod is $981.0\pm0.5$mm. The error estimate is simply based on half the smallest division. Note how we use the same number of digits after the decimal point for the measurement and its error.

The enlarged image shows that perhaps we can do a bit better. Measuring the image on a computer monitor showed that the right edge of the rod is at $981.35\pm0.05$mm. However, before we claim that level of absolute accuracy we need to think about how accurate the ruler itself is. The comparison photo of two rulers shows a discrepancy of about 0.2mm over 1000mm, which is surprisingly high. Two Toledo brand rules show agreement to within 0.05mm, so perhaps they are to be believed, however, without finding details of the not-freely available specifications we can't be sure.

Then there is the issue of thermal expansion. Stainless steel 316 has a coefficient of thermal expansion of about $1.6\times10^{-6}$/K. Rulers are generally calibrated at $20^\circ$C, so if we measure at, say, $40^\circ$C we will have a 0.032mm change in length. Not too bad. However, both the ruler and rod are made of stainless steel, so this effect should cancel out, provided that the steels are the same. Unfortunately, stainless steels show quite a range in thermal expansion, and some are half the value for 316, so the error could be as large at 0.016mm over 1000mm.

After considering all these factors, our revised reading of $981.35\pm0.05$mm is probably about right, and is a factor of ten better than our initial guess. Although if you really wanted to be sure you should look up the specs on the ruler. Certainly, $\pm0.5\text{mm}$ is too great an estimated error if you make a careful measurement.

STOP PRESS: the sleeve in which the Toledo ruler was housed says "Guaranteed Accurate to BS 4372". BS 4372 is the British Standard for metal rulers, but is not freely available, and costs about 300 dollars to buy. Some information online indicates that the ruler is likely to be guaranteed to be within 0.1mm. So, we might want to increase our error budget to that. However, a guarantee to 0.1mm is likely to mean that the standard deviation (which is the error measure we use, and where we expect 68% of rulers to be within $\pm$ one standard deviation) is considerably less than 0.1mm. Perhaps 0.05mm is a good estimate after all.

## Measurements of the resistors

Our first guess of the measurements and their errors might be $R1=215.01\pm0.01 \text{k}\Omega$ and $R2=214.66\pm0.01 \text{k}\Omega$. However, if we look up the specifications for the Agilent U1253B we find

![](http://mcba1.phys.unsw.edu.au/~mcba/u1253b-specs.png)

On the $500 \text{k}\Omega$ range the uncertainty is 0.05% of the reading plus 5 least-significant-digits. So the correct answer is $R1=215.01\pm0.16 \text{k}\Omega$ and $R2=214.66\pm0.16 \text{k}\Omega$, some 16 times higher an uncertainty as our initial guess.

As an aside, the Agilent U1253B is actually a pretty accurate handheld digital multimeter, and more commonly available ones may be considerably less accurate.

## The voltage divider

This example illustrates some of the difficulties when combining uncertainties.

We have the equation

$$V_\text{out} = V_\text{in}\times{R2\over(R1 + R2)}$$

where 

$$V_\text{in}=10.0000\pm0.0001\text{V}$$
$$R1=215.01\pm0.16 \text{k}\Omega$$
$$R2=214.66\pm0.16 \text{k}\Omega$$

If we assume that the errors in R1 and R2 are uncorrelated, then it is straightforward to calculate the uncertainty in $R1+R2$ by combining the absolute errors in quadrature. You get $R1+R2 = 429.67\pm0.23 \text{k}\Omega$. But now when you try to calculate the error in ${R2\over(R1 + R2)}$ you are faced with the problem that the error in R2 is certainly correlated to some extent with the error in $R1+R2$. We can make progress here by simply rewriting our equation for $V_\text{out}$ as

$$V_\text{out} = {V_\text{in}\over\left(1 + {R1\over R2}\right)}$$

Next we calculate the error in $R1\over R2$ using quadrature of the relative errors, and we get ${R1\over R2}=1.0016\pm0.0011$, and ${1\over\left(1 + {R1\over R2}\right)} = 0.49959\pm0.00027$ (check these calculations yourself). The error in $V_\text{in}$ is negligible by comparison, so the final result is

$$V_\text{out} = 4.9959\pm0.0027\text{V}$$

### But are the errors in R1 and R2 really uncorrelated?

In the above calculation we assumed that the errors in R1 and R2 are uncorrelated, but in reality it is likely that the main error in the multimeter is a linear scaling error, so, particularly when measuring resistances that are close to being the same value, the errors will be highly correlated. To a good approximation the uncorrelated part of the error will be equal to how stable the reading is, which is about 2 least-significant digits. So ${R1\over R2}=1.00163\pm0.00013$, which is a factor of 10 lower error than we had before. Working through the calculation we finally arrive at

$$V_\text{out} = 4.9959\pm0.0003\text{V}$$

Our initial uncertainty estimate was 10 times too high.

## Measuring the output of the voltage divider

Now we determine what the multimeter should read. Looking at the specification sheet above, we see that on the 5V range we have an error of 0.025% plus 5 digits (where the LSD is 0.0001V). So the meter should read within $\pm 0.0018\text{V}$ of the true value, and we should expect to see

$$V_\text{out} = 4.9959\pm0.0018\text{V}$$

However, things are not so simple! On page 12 of the specification we find this:

![](http://mcba1.phys.unsw.edu.au/~mcba/u1253b-specs.png)

where you can see that the multimeter has an input impedance of $10\text{M}\Omega$, which is a common value for a digital multimeter. This impedance acts in parallel across R2, making the effective resistance of the lower leg of the voltage divider equal to ${1.0\text{E}4\times R2\over1.0\text{E}4+R2}=210.14\pm0.16 \text{k}\Omega$, and the voltage we should actually read is

$$V_\text{out} = 4.9427\pm0.0018\text{V}$$

which is 30 standard deviations away from our earlier estimate!

As a final note, the resistors in this example are reasonably temperature sensitive, and the reading could be easily changed by $0.1\text{k}\Omega$ by heating the resistor with a finger. If both are at the same temperature, this effect cancels to first order.

## Measuring the lab temperature and humidity

The sensor shown is an Ecowitt model WN31C, and according to its manual the temperature uncertainty is $\pm1^\circ\text{C}$ and humidity $\pm5$%, which is really pretty good. If you look at the photo with four of these sensors, you see that the agreement between them is better than this, which doesn't necessarily mean that they are more accurate - they may all be equally wrong. Still it is impressive.

Averaging the readings from the four devices, and using the standard error on the mean, you get

$$\text{T}=21.83\pm0.03$$
$$\text{relative humidity %} = 72.5\pm0.6$$

However, these uncertainties assume that each sensor is perfectly calibrated, and we are just looking at the spread of the measurements. In reality, we have to include the manufacturer's specification of the error, and we are left with

$$\text{T}=21.8\pm1.0$$
$$\text{relative humidity %} = 73\pm5$$

Although, if we are just interested in temperature/humidity differences, then to first order we can ignore the calibration error, and our earlier estimate can be used when subtracting two measurements.
